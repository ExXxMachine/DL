```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error, mean_squared_error

# 1.  
data = pd.read_csv('XAU_15m_data_2004_to_2024-20-09.csv')
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)

# 2.  
plt.figure(figsize=(12, 6))
plt.plot(data['Close'], label=' ')
plt.title('   ')
plt.xlabel('')
plt.ylabel('')
plt.legend()
plt.show()
```

```python
# 3.   
decomposition = seasonal_decompose(data['Close'], model='additive', period=365)
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(12, 16))
decomposition.observed.plot(ax=ax1)
ax1.set_title('')
decomposition.trend.plot(ax=ax2)
ax2.set_title('')
decomposition.seasonal.plot(ax=ax3)
ax3.set_title('')
decomposition.resid.plot(ax=ax4)
ax4.set_title('')
plt.tight_layout()
plt.show()
```

```python
# 4.    ( -)
def adf_test(series):
    result = adfuller(series.dropna(), autolag='AIC')
    print(f'ADF Statistic: {result[0]}')
    print(f'p-value: {result[1]}')

adf_test(data['Close'])

# 5.   ,      ( )
if adfuller(data['Close'])[1] > 0.05:
    data_diff = data['Close'].diff().dropna()
else:
    data_diff = data['Close']
```

```python
# 6.     ARIMA
def evaluate_arima_model(X):
    train_size = int(len(X) * 0.8)
    train, test = X[:train_size], X[train_size:]
    
    best_score, best_cfg = float("inf"), None
    for p in range(3):
        for d in range(2):
            for q in range(3):
                order = (p,d,q)
                try:
                    model = ARIMA(train, order=order)
                    model_fit = model.fit()
                    forecast = model_fit.forecast(steps=len(test))
                    mae = mean_absolute_error(test, forecast)
                    rmse = np.sqrt(mean_squared_error(test, forecast))
                    if mae < best_score:
                        best_score, best_cfg = mae, order
                    print(f'ARIMA{order} MAE={mae:.3f} RMSE={rmse:.3f}')
                except:
                    continue
    
    print(f'  ARIMA{best_cfg} MAE={best_score:.3f}')
    return best_cfg

best_order = evaluate_arima_model(data_diff)
```

```python
# 7.      ARIMA
model_best = ARIMA(data_diff.dropna(), order=best_order)
model_fit_best = model_best.fit()
forecast_best = model_fit_best.forecast(steps=30)

#     
forecast_index_best = pd.date_range(start=data_diff.index[-1] + pd.Timedelta(minutes=15), periods=30, freq='15T')  #     15 
forecast_series_best = pd.Series(forecast_best.values.flatten(), index=forecast_index_best)

plt.figure(figsize=(12, 6))
plt.plot(data.index[:len(data_diff)], data_diff.values[:len(data_diff)], label='', color='blue')
plt.plot(forecast_series_best.index, forecast_series_best.values + data['Close'].iloc[-1], label=' ARIMA', color='red') #        
plt.title('  30     ARIMA')
plt.xlabel('')
plt.ylabel('')
plt.legend()
plt.show()
```
